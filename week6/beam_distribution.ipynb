# Beam Distribution

## Equation of Ellipse

The equation of an ellipse centered at the origin can be generalized as

$$
ax^2 + bxy + cy^2 = 1
$$

With simple algebraic derivations, one can easily find that the area of the ellipse is given by

$$
\mathbf{S} = \frac{2 \pi} {\sqrt{4ac - b^{2}} }
$$

Let $\mathbf{X}$ be the vector

$$
\mathbf{X} = \begin{pmatrix}
x \\
y
\end{pmatrix}
$$

and let $\mathbf{A}$ be a general $2 \times 2$ matrix

$$
\mathbf{A} =
\begin{pmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{pmatrix}
$$

With $\mathbf{X^{T}}$  being the transpose of $\mathbf{X}$, then

$$
\mathbf{X^{T} A X} =  \left( x ~~~ y \right) \left(\begin{array}{cc} a_{11} & a_{12} \\ a_{21} &  a_{22} \end{array}\right)  \left(\begin{array}{c} x \\ y \end{array}\right) 
= a_{11} x^2 + (a_{12} + a_{21}) xy + a_{22} y^2 = 1
$$

would be the equation of the ellipse in a matrix notation. If $\mathbf{A}$ is a symmetric matrix ($a_{12} = a_{21}$), then 

$$
a_{11} x^2 + (a_{12} + a_{21}) xy + a_{22} y^2 = ax^{2} + bxy + cy^{2} = 1
$$

From the previous result for the area of the ellipse, 

$$
\mathbf{S} = \frac{2 \pi} {\sqrt{4ac - b^{2}} }
= \frac{ 2 \pi } { 4a_{11}a_{22} - 4a^{2}_{12} }
= \frac{ \pi }{ \det \mathbf{A} }
$$

where $\det\mathbf{A}$ is the determinant of the symmetric matrix $\mathbf{A}$,

$$
\mathbf{A} =
\begin{pmatrix}
a_{11} & a_{12} \\
a_{12} & a_{22}
\end{pmatrix}
$$

## Particle Distribution and Covariance(Sigma) Matrix

We begin by creating a matrix $\Sigma$ made by forming the product $\mathbf{X X^{T}}$ and taking the average values across the distribution of each of its elements:

$$
\langle \mathbf{XX^{T}} \rangle 
   = \langle \left(\begin{array}{c}  x \\ y \end{array}\right) ( x \;\;\; y)\rangle
   = \left(\begin{array}{cc} \langle x^2 \rangle & \langle xy \rangle \\ \langle xy \rangle 
& \langle y^2 \rangle \end{array}\right) 
   = \left(\begin{array}{cc} \sigma_{11} & \sigma_{12} \\ \sigma_{12} & \sigma_{22} \end{array}\right) 
   \equiv \mathbf{\Sigma}
$$

Note that the inverse of $\mathbf{\Sigma}$ is

$$
\mathbf{\Sigma^{-1}} = \frac{1}{ \det\mathbf{\Sigma} } 
\left( \begin{array}{cc} \sigma_{22} & -\sigma_{12} \\ -\sigma_{12} & \sigma_{11} \end{array} \right)
$$

Then,

$$
\mathbf{X \Sigma^{-1} X^{T}} = 1
$$

This reads to the form:

$$
\sigma_{22} x^{2} - 2 \sigma_{12} xy + \sigma_{11} y^{2}
= \det \mathbf{\Sigma}
= \left( \frac{\mathbf{A}}{\pi} \right)^{2}
$$

## Twiss (Courant-Snyder, lattice functions) Parameters and Emittance

Let $\epsilon_{x}$  be the area of our relevant phase space distribution. With $y = x^{\prime} = dx/ds$ from the above derivation, then we can define new variables that related to the elements of the ellipse equation:

$$
\alpha_{x} \equiv - \frac{\pi}{ \epsilon_{x} }  \sigma_{x12}
= - \frac{\pi}{ \epsilon_{x}}  \langle x x^{\prime} \rangle
$$

$$
\beta_{x} \equiv \frac{\pi}{ \epsilon_{x} } \sigma_{x11}
= \frac{\pi}{ \epsilon_{x} }  \langle x^{2} \rangle
$$

$$
\gamma_{x} \equiv \frac{\pi}{ \epsilon_{x} }  \sigma_{x22}
= \frac{\pi}{ \epsilon_{x} }  \langle x^{\prime 2} \rangle
$$

The **Sigma Matrix** is then
$$
\mathbf{\Sigma} = 
\begin{pmatrix} 
\langle x^2\rangle & \langle xx^{\prime}\rangle\\
\langle xx^{\prime}\rangle & \langle x^{\prime2}\rangle 
\end{pmatrix} = 
\begin{pmatrix} 
\beta_{x}\epsilon_{x} & -\alpha_{x}\epsilon_{x}\\
-\alpha_{x}\epsilon_{x} & \gamma_{x}\epsilon_{x}
\end{pmatrix}
$$

Then, the equation of the phase space ellipse can be written as:

$$
\gamma_{x} x^{2} + 2 \alpha_{x} x x^{\prime} + \beta_{x} x^{\prime 2} = \epsilon / \pi
$$

The parameters, $\alpha_{x}$, $\beta_{x}$, and $\gamma_{x}$ are referred to as the **Twiss** parameters, (or **Courant-Snyder** parameters, **lattice functions**)
  are referred to as the Courant-Snyder parameters

The Twiss parameters represent the orientation and eccentricity of the phase space ellipse that mimics the orientation and shape of the particle distribution, while the size of the ellipse is governed by $\epsilon_{x}$. The area in the phase space, $\epsilon_{x}$, containing the particles is called the **emittance**.

The emittance defined by this ellipse is often called the *rms emittance* and it is easy to verify that the rms emittance is given by

$$
\epsilon_{x} = \pi \sqrt{ \langle x^{2} \rangle \langle x^{\prime2} \rangle - \langle x x^{\prime} \rangle^{2} }
$$

Note that

$$
\beta_{x}\gamma_{x}-\alpha_{x}^2 = [\sigma_{x11}/(\epsilon_{x}/\pi)]\;[\sigma_{x22}/(\epsilon_{x}/\pi)] - 
   [ \sigma_{x12}/(\epsilon_{x}/\pi)]^2 = \det\mathbf{\Sigma}/(\epsilon_{x}/\pi)^2 = 1.
$$

and then

$$
\gamma_{x} = \frac{ 1 + \alpha^{2}_{x} }{ \beta_{x} }
$$

## Extension to Multiple Dimensions

In particle accelerators with linear elements (drifts, dipoles, and quadrupoles, etc.), the motion of particles in the one (say, $x$) direction is independent of the other ($y$). 

For such a system, one can analyze the motion in $x$ using $2 \times 2$ matrices describing trajectory development through each independent element and simultaneously use appropriate $2 \times 2$ matrices to describe the motion in $y$ through the same elements. 

Then, independent Twiss parameters for each direction can be used. 

For uncoupled motion, the vector

$$
\mathbf{X} = \begin{pmatrix}
x \\
x^{\prime} \\
y \\
y^{\prime}
\end{pmatrix}
$$

can be operated on a general $4 \times 4$ matrix

$$
\mathbf{M} = \left( \begin{array}{cccc}
   \sigma_{11} & \sigma_{12} & 0 & 0 \\ 
   \sigma_{21} & \sigma_{22} & 0 & 0 \\
    0     &   0 & \sigma_{33} & \sigma_{34} \\
    0     &   0 & \sigma_{43} & \sigma_{44}
   \end{array}\right)
$$

But in the case that the particle motion is coupled between the four phase space variables, such as in the case of transport through a solenoidal magnetic field, or where ideal quadrupole fields are rotated about the longitudinal beam axis, the traditional Twiss description is no longer valid.

However, our matrix treatment using a $\mathbf{\Sigma}$ matrix  is indeed still valid. The determinant of a $4 \times 4$ matrix will still be an invariant of the motion, and the matrix elements will still propagate according to

$$
\mathbf{\Sigma_{f}} = \mathbf{M \Sigma_{i} M^{T}}
$$


## Numerical Examples

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Calculate Twiss Parameters

```python
x, xp = np.random.normal(3, 2.5, size=(2, 10000))
```

```python
fig = plt.figure(figsize=(10,10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:,:3])
x_hist = fig.add_subplot(grid[0,:3], yticklabels=[], sharex=main_ax)
y_hist = fig.add_subplot(grid[1:,3], xticklabels=[], sharey=main_ax)

# scatter points on the main axes
main_ax.scatter(x, xp, s=1)#, 'ob', markersize=3, alpha=0.2)
main_ax.set_xlabel(r'$x$', fontsize=30)
main_ax.set_ylabel(r'$x^{\prime}$', fontsize=30)
main_ax.grid()

# histogram on the attached axes
x_hist.hist(x, 100, histtype='step', orientation='vertical', color='b')
y_hist.hist(xp, 100, histtype='step', orientation='horizontal', color='b')
        
plt.show()
```

```python
sig11 = np.mean(x*x)
sig12 = np.mean(x*xp)
sig22 = np.mean(xp*xp)
```

```python
epsilon_x = np.sqrt(sig11*sig22-sig12**2)
print (epsilon_x)
```

```python
alpha_x = - sig12 / epsilon_x
beta_x = sig11 / epsilon_x
gamma_x = sig22 / epsilon_x
print (alpha_x, beta_x, gamma_x)
print ((1 + alpha_x**2) / beta_x)
```

### Generate Beam Distribution from Twiss Parameters

```python
ax = -0.5
bx = 1.3
cx = 2.1
ex = 3.1
```

```python
s11 = ex * bx
s12 = - ex * ax
s22 = ex * cx
cov = np.zeros((2,2))
cov[0,0] = s11
cov[0,1] = s12
cov[1,0] = s12
cov[1,1] = s22
```

```python
mean = [np.sqrt(s11), np.sqrt(s22)]
Np = 10000
newx, newxp = np.random.multivariate_normal(mean, cov, Np).T
```

```python
fig = plt.figure(figsize=(10,10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:,:3])
x_hist = fig.add_subplot(grid[0,:3], yticklabels=[], sharex=main_ax)
y_hist = fig.add_subplot(grid[1:,3], xticklabels=[], sharey=main_ax)

# scatter points on the main axes
main_ax.scatter(newx, newxp, s=1)#, 'ob', markersize=3, alpha=0.2)
main_ax.set_xlabel(r'$x$', fontsize=30)
main_ax.set_ylabel(r'$x^{\prime}$', fontsize=30)
main_ax.grid()

# histogram on the attached axes
x_hist.hist(newx, 100, histtype='step', orientation='vertical', color='b')
y_hist.hist(newxp, 100, histtype='step', orientation='horizontal', color='b')
        
plt.show()
```

### Generate Beam Distribution with Dispersion

After passing through bending elements, the trajectories of the various particles become dispersed by the amount $D\delta$, where $D$ is the dispersion function and $\delta = \Delta p / p$ is the momentum spread relative to a reference particle. Then, the particle displacement can be written as

$$
x = x_{\beta} + D_{x}\delta_{x}
$$

By averaging over the particle distribution:

$$
\begin{eqnarray*}
\langle x^2 \rangle &=& \langle x_\beta^2\rangle + \langle x_\beta\;D_{x}\delta_{x}\rangle +\langle D_{x}^2\delta_{x}^2\rangle  = \epsilon_{x}\beta + D^2\langle\delta_{x}^2\rangle \\
\langle x^{\prime2}\rangle &=& \epsilon_{x}\gamma_{x} + D_{x}^{\prime2}\langle\delta_{x}^2\rangle \\
\langle x^{\prime}x\rangle &=& -\epsilon_{x}\alpha_{x} + D_{x}^{\prime}D\langle\delta_{x}^2\rangle \\
\langle x\cdot\delta_{x}\rangle &=& \langle x_\beta\delta_{x}\rangle + \langle D_{x}\delta_{x}^2\rangle = D_{x}\langle\delta_{x}^2\rangle \\
\langle x^{\prime}\cdot\delta_{x}\rangle &=& \langle x_\beta^{\prime}\delta_{x}\rangle + \langle D_{x}^{\prime}\delta_{x}^2\rangle = D_{x}^{\prime}\langle\delta_{x}^2\rangle
\end{eqnarray*}\
$$

Then we can find the values of the dispersion function, Twiss parameters and emittance that describe the particle distribution:

$$
\begin{eqnarray*}
D_{x} &=& \frac{\langle x\;\delta_{x}\rangle}{\langle\delta_{x}^2\rangle} \\
D_{x}^{\prime} &=& \frac{\langle x^{\prime}\delta_{x}\rangle}{\langle\delta_{x}^2\rangle} \\
\epsilon_{x}\beta_{x} &=& \langle x^2\rangle - \frac{\langle x\delta_{x}\rangle^2}{\langle\delta_{x}^2\rangle} \\
\epsilon_{x}\gamma_{x} &=& \langle x^{\prime2}\rangle - \frac{\langle x^{\prime}\delta_{x}\rangle^2}{\langle\delta_{x}^2\rangle} \\
\epsilon_{x}\alpha_{x} &=& -\langle xx^{\prime}\rangle + \frac{\langle x^{\prime}\delta_{x}\rangle\langle x\;\delta_{x}\rangle}{\langle\delta_{x}^2\rangle} \\
\epsilon_{x} &=& \sqrt{(\epsilon_{x}\beta_{x})(\epsilon_{x}\gamma_{x})-(\epsilon_{x}\alpha_{x})^2}
\end{eqnarray*}
$$

```python
ax = -2.5
bx = 17.0
cx = (1 + ax * ax) / bx
ex = 12.0
Dx =  1.4
Dxp =  0.015
Np = 20000
```

```python
delta = np.mean(np.random.normal(0, 1, Np))*11.0
```

```python
cov = np.zeros((6,6))
cov[0,0] = ex * bx + (Dx * delta)**2
cov[0,1] = cov[1,0] = ex * ax + Dx * Dxp * delta**2
cov[1,1] = ex * cx + (Dxp + delta)**2
cov[0,5] = cov[5,0] = Dx * delta**2
cov[1,5] = cov[5,1] = Dxp * delta**2
```

```python
mean = [np.sqrt(cov[0,0]), np.sqrt(cov[1,1]), 0, 0, 0, 0]
particles = np.random.multivariate_normal(mean, cov, Np)
```

```python
x = particles[:,0]
xp = particles[:,1]
```

```python
fig = plt.figure(figsize=(10,10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:,:3])
x_hist = fig.add_subplot(grid[0,:3], yticklabels=[], sharex=main_ax)
y_hist = fig.add_subplot(grid[1:,3], xticklabels=[], sharey=main_ax)

# scatter points on the main axes
main_ax.scatter(x, xp, s=1)#, 'ob', markersize=3, alpha=0.2)
main_ax.set_xlabel(r'$x$', fontsize=30)
main_ax.set_ylabel(r'$x^{\prime}$', fontsize=30)
main_ax.grid()

# histogram on the attached axes
x_hist.hist(x, 100, histtype='step', orientation='vertical', color='b')
y_hist.hist(xp, 100, histtype='step', orientation='horizontal', color='b')
        
plt.show()
```
