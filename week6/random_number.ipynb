## Random Number

### Random number generator and pseudorandom number generator

In our simulations and data analysis, we often use random number generatom. A random number generator is a system that generates random numbers from a **true** source of randomness. Pseudorandomness is a sample of numbers that look close to random, but were generated using a deterministic process. The source of randomness that we are using in the code mostly pseudorandom number generator

The numbers are generated in a sequence, which is determined by an initial number, called **seed**. If you do not explicitly choose a seed, then it may use the current system time in seconds or milliseconds as the seed. The value of the seed does not matter. If you use the same seed number for repeated processes, then the generator will return the same sequence of random numbers.

#### Gaussian (Normal) Distribution

The probability density function of the normal distribution was first derived by De Moivre and 200 years later by both Gauss and Laplace independently.

$$
f(x) = \frac{1}{2 \pi \sigma^{2}} \exp \left[ - \frac{ (x - \mu)^{2} }{ 2 \sigma^{2} } \right]
$$

```python
mean = 0
sigma = 0.1
size = 10000
```

```python
x = np.random.normal(mean, sigma, size)
```

Verify the numbers

```python
print (np.mean(x), np.std(x))
```

```python
count, bins, ignored = plt.hist(x, 50, density=True, histtype='step')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mean)**2 / (2 * sigma**2) ), linewidth=1, color='r')
plt.show()
```

#### N-dimensional random samples

```python
x, y = np.random.normal(3, 2.5, size=(2, 10000))
```

```python
fig = plt.figure(figsize=(10,10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:,:3])
x_hist = fig.add_subplot(grid[0,:3], yticklabels=[], sharex=main_ax)
y_hist = fig.add_subplot(grid[1:,3], xticklabels=[], sharey=main_ax)

# scatter points on the main axes
main_ax.scatter(x, y, s=1)#, 'ob', markersize=3, alpha=0.2)
main_ax.set_xlabel(r'$x$', fontsize=30)
main_ax.set_ylabel(r'$y$', fontsize=30)
main_ax.grid()

# histogram on the attached axes
x_hist.hist(x, 100, histtype='step', orientation='vertical', color='b')
y_hist.hist(y, 100, histtype='step', orientation='horizontal', color='b')
        
plt.show()
```

```python
mean = [0, 0]
cov = [[10, 1], [1, 10]]
Np = 10000
x, y = np.random.multivariate_normal(mean, cov, Np).T
```

```python
fig = plt.figure(figsize=(10,10))

grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[1:,:3])
x_hist = fig.add_subplot(grid[0,:3], yticklabels=[], sharex=main_ax)
y_hist = fig.add_subplot(grid[1:,3], xticklabels=[], sharey=main_ax)

# scatter points on the main axes
main_ax.scatter(x, y, s=1)#, 'ob', markersize=3, alpha=0.2)
main_ax.set_xlabel(r'$x$', fontsize=30)
main_ax.set_ylabel(r'$y$', fontsize=30)
main_ax.grid()

# histogram on the attached axes
x_hist.hist(x, 100, histtype='step', orientation='vertical', color='b')
y_hist.hist(y, 100, histtype='step', orientation='horizontal', color='b')
        
plt.show()
```
