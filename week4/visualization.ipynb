### Plotting more than one curve

````python
def xcubic(x):
    return x*x*x
    
y2 = xcubic(x)
y3 = x**4
y4 = np.exp(x)
````

````python
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.title(r'2D plot examples', fontsize=16)
plt.plot(x, y, "bs-")
plt.plot(x, y2, "gd:")
plt.plot(x, y3, "cv-.")
plt.plot(x, y4, "m^--")
plt.grid(1)
plt.show()
````

### 3D parametric plot

Consider the parameterized data set $t$ is a vector from 0 to $4\pi$ with a step $\pi/20$, $x = \sin(t)$, and $y = \cos(t)$. Make a three-dimensional plot of the $(x,y,t)$ data set using plot3. Turn the grid on, make the axis equal, and put axis labels and a title. 

````python
pi = np.pi
t = np.arange(0, 4*pi, pi/20)
x = np.sin(t)
y = np.cos(t)

fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()

ax.plot3D(x, y, t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=16)
ax.set_ylabel('y', labelpad=16)
ax.set_zlabel('t', labelpad=16)

plt.show()
````

### Surface plot

In three-dimensional surface plotting, we wish to make a graph of some relationship $f(x, y)$. In surface plotting all $(x,y)$ pairs must be given. This is not straightforward to do using vectors. Therefore, in surface plotting, the first data structure you must create is called a mesh. 

````python
x = [1, 2, 3]
y = [4, 5, 6]

X, Y = np.meshgrid(x, y)
print(X)
print(Y)
````

The function to plot the 3D surfaces is plot_surface(X,Y,Z), where $X$ and $Y$ are the output arrays from meshgrid, and $Z=f(X,Y)$ or $Z(i,j)=f(X(i,j),Y(i,j))$. The most common surface plotting functions are surf and contour.

Make a plot of the surface $f(x,y)=cos(x)*sin(y)$ for $−\pi \le x \le \pi$, and $−\pi \le y \le \pi$ using the plot_surface function.

````python
x = np.arange(-pi, pi, 0.1)
y = np.arange(-pi, pi, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.sin(Y)
````

````python
fig = plt.figure(figsize = (16,16))
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.plasma)

ax.set_xlabel('x', labelpad=16, fontsize=20)
ax.set_ylabel('y', labelpad=16, fontsize=20)
ax.set_zlabel('z', labelpad=16, fontsize=20)

fig.colorbar(surf, shrink=0.7)

plt.show()
````

### Wireframe plot

````python
fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.plot_wireframe(X,Y,Z)
ax1.set_title('Wireframe plot')

ax1.set_xlabel('x', labelpad=16, fontsize=20)
ax1.set_ylabel('y', labelpad=16, fontsize=20)
ax1.set_zlabel('z', labelpad=16, fontsize=20)


ax2 = fig.add_subplot(1, 3, 2, projection='3d')
surf1 = ax2.plot_surface(X, Y, Z, cmap = plt.cm.plasma)

ax2.set_xlabel('x', labelpad=16, fontsize=20)
ax2.set_ylabel('y', labelpad=16, fontsize=20)
ax2.set_zlabel('z', labelpad=16, fontsize=20)

ax2.set_title('Surface plot')

# Normalize to [0,1]
norm = plt.Normalize(Z.min(), Z.max())
colors = plt.cm.plasma(norm(Z))
rcount, ccount, _ = colors.shape

ax3 = fig.add_subplot(1, 3, 3, projection='3d')
surf2 = ax3.plot_surface(X, Y, Z, rcount=rcount, ccount=ccount,
                       facecolors=colors, shade=False)
surf2.set_facecolor((0,0,0,0))
ax3.set_title('Wireframe plot')

ax3.set_xlabel('x', labelpad=16, fontsize=20)
ax3.set_ylabel('y', labelpad=16, fontsize=20)
ax3.set_zlabel('z', labelpad=16, fontsize=20)

plt.tight_layout()

plt.show()
````

````python
n = 100
x = np.linspace(0, 2*pi, n)
y = np.sin(x)

#metadata
FFMpegWriter = manimation.writers['ffmpeg']
metadata = dict(title='Particle Animation')
writer = FFMpegWriter(fps=15, metadata=metadata)

#plots
fig = plt.figure()
wave, = plt.plot(x, y, 'b')
particle, = plt.plot([], [], 'ro', markersize = 5)
plt.xlabel('x')
plt.ylabel('sin(x)')

# Saving the frames for the movie
with writer.saving(fig, "particle.mp4", 100):
    for i in range(n):
        x0 = x[i]
        y0 = y[i]
        particle.set_data(x0, y0)
        writer.grab_frame()

plt.show()
````
